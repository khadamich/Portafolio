# Feature Engineering

In this notebook I'm gonna prepare the dataset for machine learning models

```python
import pandas as pd
```

```python
madrid_dataset = pd.read_pickle("01_data_interim.pkl")
```

```python
madrid_dataset
```

## Changing results to numerical values

```python
#The target variable
res_val = []
for i in range(len(madrid_dataset)):
    if madrid_dataset['Result'].iloc[i] == 'W':
        res_val.append(2)
    elif madrid_dataset['Result'].iloc[i] == 'D':
        res_val.append(1)
    else:
        res_val.append(0)
if "res_val" not in madrid_dataset.columns:
        madrid_dataset.insert(17, "res_val", res_val, True)
```

```python
res_val_ht = []
for i in range(len(madrid_dataset)):
    if madrid_dataset['Result_HT'].iloc[i] == 'W':
        res_val_ht.append(2)
    elif madrid_dataset['Result_HT'].iloc[i] == 'D':
        res_val_ht.append(1)
    else:
        res_val_ht.append(0)
if "res_val_ht" not in madrid_dataset.columns:
        madrid_dataset.insert(17, "res_val_ht", res_val_ht, True)
```

```python
#Drop the unneeded features
madrid_dataset = madrid_dataset.drop('Result', axis = 1)
madrid_dataset = madrid_dataset.drop('Result_HT', axis = 1)
madrid_dataset = madrid_dataset.drop('Points', axis = 1)
```

## Make dummies for rivals

```python
madrid_dataset.head()
```

```python
features = pd.get_dummies(madrid_dataset, columns=['Rival']).astype(int) 
features.head()
```

```python
column_to_move = features.pop("res_val")

# insert column with insert(location, column_name, column_value)

features.insert(0, "Result", column_to_move)
```
